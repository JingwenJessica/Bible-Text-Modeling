```python
#!/usr/bin/env python
"""
Masked wordcloud
================
Using a mask you can generate wordclouds in arbitrary shapes.
"""

from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS

%matplotlib inline
```

```python
# d = path.dirname(__file__)

# Read the whole text.
# text = open('bible_king.txt').read()
# text = open('bible_king_oldT.txt').read()
text = open('bible_king_newT.txt').read()
```

```python
# read the mask image
# taken from
# http://www.stencilry.org/stencils/movies/alice%20in%20wonderland/255fk.jpg
# bible_mask = np.array(Image.open(path.join(d, "bible_mask.png")))
# bible_mask = np.array(Image.open("bible_mask.png"))

stopwords = set(STOPWORDS)
stopwords.add("shall")
stopwords.add("unto")
stopwords.add("will")
stopwords.add("may")
stopwords.add("even")
stopwords.add("let")
stopwords.add("might")
# stopwords.add("day")

# wc = WordCloud(background_color="white", max_words=2000, mask=alice_mask, stopwords=stopwords)
# wc = WordCloud(background_color="white", max_words=2000, max_font_size=100, stopwords=stopwords)
wc = WordCloud(background_color="white", max_words=1000, stopwords=stopwords)

# generate word cloud

wc.generate(text)

# store to file
# wc.to_file("bible_king.png")
# wc.to_file("bible_king_oldT.png")
wc.to_file("bible_king_newT.png")
```

```python
# show
plt.imshow(wc)
plt.axis("off")
plt.figure()
# plt.imshow(bible_mask, cmap=plt.cm.gray)
# plt.axis("off")
plt.show()
```
